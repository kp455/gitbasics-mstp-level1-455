# matplotlib ticks
> ticks is values within a coordinates

```python
import matplotlib.pyplot as pl
y=[5,10,15,20,25,30,35,40,45,50]
print(y)
x=[9,8,5,6,3,2,1,4,7,10]
print(x)
pl.plot(x,y)
pl.title('A graph with a example to ticks')
pl.xlabel('Natural numbers')
pl.ylabel('values with ++5 up to 50')
pl.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.title('Matplotlib-Ticks')
pl.xlabel('10 Natural numbers')
pl.ylabel('5-50 values with 5+')
pl.xticks(np.arange(10,20,1))
pl.yticks(np.arange(5,51,5))
pl.plot(x,y)
print(x)
print(y)
pl.show()

```

# Representing a sine wave

. A=sine(wt)A=Amplitude
.  w=frequency
.  t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.title("sine wave A=sin(wt)")
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
pl.grid()
pl.plot(years,Alaska_income)
pl.title('Income of Alaska 2005 to 2013')
pl.xlabel('years')
pl.ylabel('Alaska_Income')
pl.savefig('Income of Alaska 2005 to 2013.jpg')
pl.show()
```

### All states income from 2005 to 2013

```python
import matplotlib.pyplot as pl 
import pandas as pd
years=df.columns[-5::]
Allstatesincome=df.iloc[0:,2:]
pl.grid()
pl.plot(years,Allstatesincome)
pl.title('All states of income 2005 to 2013')
pl.xlabel('years')
pl.ylabel('All statesincome')
pl.savefig('All states of income 2005 to 2013.jpg')
pl.show()
```
